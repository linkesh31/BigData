```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets
df_orders = pd.read_csv('data/amazon_products.csv')
# Load the dataset with low_memory=False to handle mixed types
df_sales = pd.read_csv('data/Amazon Sales FY2020-21.csv', low_memory=False)

df_categories = pd.read_csv('data/amazon_categories.csv')
df_amazingmart = pd.read_excel('data/P1-AmazingMartEU2.xlsx')  # Adjust path if needed

# Display the first few rows of each dataset for a quick overview
print("Orders Dataset:")
display(df_orders.head())
print("Sales Dataset:")
display(df_sales.head())
print("Categories Dataset:")
display(df_categories.head())
print("AmazingMart Dataset:")
display(df_amazingmart.head())
```

```python
# Display basic information and check for missing data in each dataset
datasets = {
    "Orders Dataset": df_orders,
    "Sales Dataset": df_sales,
    "Categories Dataset": df_categories,
    "AmazingMart Dataset": df_amazingmart
}

# Loop through each dataset to show info and missing values
for name, df in datasets.items():
    print(f"\n{name} Info:")
    df.info()
    print("\nMissing Values:")
    print(df.isnull().sum())
```

```python
# Fill missing values for numeric columns only
numeric_columns_orders = df_orders.select_dtypes(include=['number']).columns
numeric_columns_sales = df_sales.select_dtypes(include=['number']).columns
numeric_columns_amazingmart = df_amazingmart.select_dtypes(include=['number']).columns

# Fill numeric columns with the median value
df_orders[numeric_columns_orders] = df_orders[numeric_columns_orders].fillna(df_orders[numeric_columns_orders].median())
df_sales[numeric_columns_sales] = df_sales[numeric_columns_sales].fillna(df_sales[numeric_columns_sales].median())
df_amazingmart[numeric_columns_amazingmart] = df_amazingmart[numeric_columns_amazingmart].fillna(df_amazingmart[numeric_columns_amazingmart].median())

# Fill missing values in non-numeric columns with 'Unknown'
df_orders.fillna('Unknown', inplace=True)
df_sales.fillna('Unknown', inplace=True)
df_categories.fillna('Unknown', inplace=True)
df_amazingmart.fillna('Unknown', inplace=True)

# Verify that missing values have been handled
datasets = {
    "Orders Dataset": df_orders,
    "Sales Dataset": df_sales,
    "Categories Dataset": df_categories,
    "AmazingMart Dataset": df_amazingmart
}

for name, df in datasets.items():
    print(f"\n{name} Missing Values After Handling:")
    print(df.isnull().sum())
```

```python
# Remove duplicate rows in each dataset
for name, df in datasets.items():
    initial_count = df.shape[0]  # Initial number of rows
    df.drop_duplicates(inplace=True)
    final_count = df.shape[0]  # Number of rows after removing duplicates
    print(f"\n{name}: Removed {initial_count - final_count} duplicate rows.")

# Verify that duplicates have been removed
for name, df in datasets.items():
    print(f"{name} - Duplicates Remaining: {df.duplicated().sum()}")
```

```python
# Display column names to help identify potential date columns
print("Column names in Sales Dataset:")
print(df_sales.columns)
print("\nColumn names in AmazingMart Dataset:")
print(df_amazingmart.columns)

# Automatically identify columns with 'date' in their names and convert them to datetime format
for column in df_sales.columns:
    if 'date' in column.lower():
        df_sales[column] = pd.to_datetime(df_sales[column], errors='coerce')
        print(f"Converted '{column}' to datetime in Sales Dataset.")

for column in df_amazingmart.columns:
    if 'date' in column.lower():
        df_amazingmart[column] = pd.to_datetime(df_amazingmart[column], errors='coerce')
        print(f"Converted '{column}' to datetime in AmazingMart Dataset.")

# Check the updated data types for verification
print("\nSales Dataset Data Types After Conversion:")
print(df_sales.dtypes)
print("\nAmazingMart Dataset Data Types After Conversion:")
print(df_amazingmart.dtypes)
```

```python
# Example: Encoding categorical variables using label encoding
# This will convert categorical columns to numeric format for analysis

# List of columns to encode in the Sales dataset
categorical_columns_sales = ['status', 'category', 'payment_method', 'Region']

# Apply label encoding using pandas' .astype('category').cat.codes
for column in categorical_columns_sales:
    df_sales[column + '_encoded'] = df_sales[column].astype('category').cat.codes

# List of columns to encode in the AmazingMart dataset
categorical_columns_amazingmart = ['Customer Name', 'City', 'Country', 'Region', 'Segment', 'Ship Mode']

# Apply label encoding
for column in categorical_columns_amazingmart:
    df_amazingmart[column + '_encoded'] = df_amazingmart[column].astype('category').cat.codes

# Display the first few rows to check the encoded columns
print("\nEncoded Columns in Sales Dataset:")
display(df_sales.head())

print("\nEncoded Columns in AmazingMart Dataset:")
display(df_amazingmart.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# List of columns to visualize for outliers
columns_to_check = ['price_x', 'qty_ordered', 'total']

# Plot boxplots for each column
for column in columns_to_check:
    plt.figure(figsize=(10, 5))
    sns.boxplot(x=df_final_combined[column])
    plt.title(f'Boxplot for {column}')
    plt.show()
```

```python
# Capping outliers using the IQR method
for column in columns_to_check:
    Q1 = df_final_combined[column].quantile(0.25)
    Q3 = df_final_combined[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Cap the outliers
    df_final_combined[column] = df_final_combined[column].apply(lambda x: upper_bound if x > upper_bound else (lower_bound if x < lower_bound else x))
```

```python
# Save the final preprocessed dataset to a CSV file
output_path = 'data/preprocessed_combined_dataset.csv'
df_final_combined.to_csv(output_path, index=False)

print(f"Preprocessed dataset has been saved to: {output_path}")
```
